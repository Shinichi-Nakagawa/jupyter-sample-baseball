```python
import pandas as pd
import sqlite3
pd.options.display.max_columns = 40
pd.options.display.max_rows = 50
```

```python
con = sqlite3.connect('baseball.db')
```

```python
# まずは普通にクエリを叩く
results = con.execute('select name, team, ba, hr, rbi from batter  where pa >= 300 order by ba desc')
```

```python
for row in results:
    print(row)
```

```python
# 続いてpandasで読み込む
df_batter = pd.read_sql('select * from batter', con)
```
