```python
# 必要なライブラリをimport
import pandas as pd
import sqlite3

# notebook内のカラム数と行数の設定(割と適当)
pd.options.display.max_columns = 50
pd.options.display.max_rows = 100
```

```python
# コネクションを作る, 引数はsqlite3ファイルまでのパス(相対でも絶対でもOK)
con = sqlite3.connect('baseball.db')
```

```python
# まずは普通にクエリを叩く(両リーグで300打席以上の出場者の名前・チーム・打率・打点・本塁打を打率の高い順に)
results = con.execute('select name, team, ba, pa, hr, rbi from batter  where pa >= 300 order by ba desc limit 30')
```

```python
for row in results:
    print(row)
```

```python
# 続いてpandasで読み込む(全打撃データ)
df_batter = pd.read_sql('select * from batter', con)
```

```python
# 中身をチェック
df_batter
```

```python
# さっきのSQLと同じようなことをやってみる(上位20名)
df_batter[['name', 'team', 'pa', 'ba', 'hr', 'rbi']].where(df_batter['pa'] >= 300).sort_values('ba', ascending=False).head(30)
```

```python
# この辺からpandasの機能で遊ぶ,まずは基本統計量
df_batter.describe().round(1)
```

```python
# チームごとにgroupbyしてみてもう一度統計量を取る
df_batter_grouped = df_batter.groupby('team')
```

```python
df_batter_grouped.mean().round(1)
```

```python
df_batter_grouped.describe().round(1)
```

```python
df_batter_grouped.max()
```

```python
# 主要指標の分布をhistogramで出す

import matplotlib.pyplot as plt
```

```python
# 打率
ax = df_batter['ba'].where(df_batter['pa'] > 400).hist(bins=20)
ax.set_title('histogram by batting average')
plt.show()
```

```python
# 出塁率
ax = df_batter['obp'].where(df_batter['pa'] > 400).hist(bins=20)
ax.set_title('histogram by On Base')
plt.show()
```

```python
# 長打率
ax = df_batter['slg'].where(df_batter['pa'] > 400).hist(bins=20)
ax.set_title('histogram by Slugging')
plt.show()
```

```python
# 最後に,セイバーメトリクス指標で最も簡単なOPS(On Base Plus Slugging)を求める
df_batter['ops'] = df_batter['obp'] + df_batter['slg']
```

```python
# OPSランキング
df_batter[['name', 'team', 'pa', 'ba', 'hr', 'rbi', 'obp', 'slg', 'ops']].where(df_batter['pa'] >= 400).sort_values('ops', ascending=False).head(30)
```

```python
# OPSの分布
ax = df_batter['ops'].where(df_batter['pa'] > 400).hist(bins=20)
ax.set_title('histogram by OPS')
plt.show()
```
